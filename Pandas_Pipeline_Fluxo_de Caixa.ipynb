```python
# 2.4 Desnormalizar os dados do arquivo parquet para tabelas Dimensões e Fato
#      2.4.1 - Tabelas Dimensões: Pessoas( cod_clifor, nomeclifor), CondicaoPagamento( condicao_pagamento) e TipoDoc( tipo_doc)

# Criando as tabelas dimensões: Pessoas | Condição de Pagamento | Tipo de documento
# Dimensão - Pessoas
df_pessoas = df_silver_pessoas[['cod_clifor', 'nomeclifor']].drop_duplicates() 
# Dimensão - Condicao de Pagamento
df_condicao_pagamento = df_silver_fcx[['condicao_pagamento']].drop_duplicates() 
# Dimensão - Tipo de documento
df_tipo_doc = df_silver_fcx[['tipo_doc']].drop_duplicates()
```

```python
# 2.4 Desnormalizar os dados do arquivo parquet para tabelas Dimensões e Fato
#      2.4.2 -  Tabela Fato : Fato_fcx ( documento, cod_clifor,tipo_doc , valor, data_emissão, data_vencimento, data_pagamento condicao_pagamento)

# Criando a Tabela Fato: Fluxo de Caixa
df_fato_fcx = df_silver_fcx[['documento', 'cod_clifor', 'tipo_doc', 'valor', 'data_emissão', 'data_vencimento', 'data_pagamento', 'condicao_pagamento']]
# Exibindo Tabela Fato
df_fato_fcx
```

```python
# 2.5 - Salvar na camada Silver
#      2.5.1 - Criar Dimensão Pessoas


# Define silver_path e subpasta
silver_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\silver"
subpasta = 'pessoas'
nome_arquivo = 'pessoas.parquet'

# Define o caminho para salvar o arquivo pessoas
pessoas_path = os.path.join(silver_path, subpasta, nome_arquivo)

# Cria a pasta se necessário
os.makedirs(os.path.dirname(pessoas_path), exist_ok=True)

# Salva o dataframe no caminho especificado
df_pessoas.to_parquet(pessoas_path, index=False)

print(f"Arquivo salvo em: {pessoas_path}")
```

```python
# 2.5 - Salvar na camada Silver
#      2.5.2 - Criar Dimensão Condição de Pagamento


# Define silver_path e subpasta
silver_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\silver"
subpasta = 'condicao_pagamento'
nome_arquivo = 'condicao_pagamento.parquet'

# Define o caminho para salvar o arquivo Condição de Pagamento
cp_path = os.path.join(silver_path, subpasta, nome_arquivo)

# Cria a pasta se necessário
os.makedirs(os.path.dirname(cp_path), exist_ok=True)

# Salva o dataframe no caminho especificado
df_condicao_pagamento.to_parquet(cp_path, index=False)

print(f"Arquivo salvo em: {cp_path}")
```

```python
# 2.5 - Salvar na camada Silver
#      2.5.3 - Criar Dimensão Tipo de Documento


# Define silver_path e subpasta
silver_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\silver"
subpasta = 'tipo_doc'
nome_arquivo = 'tipo_doc.parquet'

# Define o caminho para salvar o arquivo Tipo de Documento
doc_path = os.path.join(silver_path, subpasta, nome_arquivo)

# Cria a pasta se necessário
os.makedirs(os.path.dirname(doc_path), exist_ok=True)

# Salva o dataframe no caminho especificado
df_tipo_doc.to_parquet(doc_path, index=False)

print(f"Arquivo salvo em: {doc_path}")
```

```python
# 2.5 - Salvar na camada Silver
#      2.5.4 - Criar Fato: fluxo de Caixa
#              2.5.4.1  - Garantir formato de dados correto, como no campo Data
#              2.5.4.2  - Particionar a data em Ano/Mes

# Define silver_path e subpasta
silver_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\silver"
subpasta = 'transacoes'
pasta_completa = os.path.join(silver_path, subpasta)

# Configurar o formato da coluna 'data_emissão' para o formato datetime
df_fato_fcx['data_emissão'] = pd.to_datetime(df_fato_fcx['data_emissão'])

# Criar novas colunas: Ano e Mês
df_fato_fcx['Ano'] = df_fato_fcx['data_emissão'].dt.year
df_fato_fcx['Mes'] = df_fato_fcx['data_emissão'].dt.month

# Salvar o arquivo parquet particionado por Ano/Mes automaticamente
df_fato_fcx.to_parquet(
    pasta_completa,
    index=False,
    partition_cols=['Ano', 'Mes'],
    engine='pyarrow'  # verifica se o PyArrow está instalado
)

print(f"Arquivo salvo particionado em: {pasta_completa}")
```

```python
# 3.1 - Dimensão Pessoas
#       3.1.1 - Definir caminhos de leitura e gravação de arquivos relacionados as camadas Silver e Gold
#       3.1.2 - Criar Dimensão Pessoas na camada Gold
#       3.1.3 - Criar Surrogate Key para Dimensão Pessoas
#       3.1.4 - Salvar Dimensão Pessoas em formato parquet

# Define caminhos 
gold_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\gold" 
silver_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\silver"

# Lê arquivo parquet da camada Silver
tabela = 'pessoas'
pasta_completa = os.path.join(silver_path, tabela)
df_pessoas = pd.read_parquet(os.path.join(pasta_completa)) 

# Cria chaves substitutas (surrogate keys) 
df_pessoas['pessoa_sk'] = range(1, len(df_pessoas) + 1)

# Salva a dimensão na camada gold 
subpasta = 'pessoas'
nome_arquivo = 'pessoas.parquet'
pasta_completa = os.path.join(gold_path, tabela, nome_arquivo)

# Cria diretórios caso não existam 
os.makedirs(os.path.dirname(pasta_completa), exist_ok=True)

# Salva Arquivo Parquet na Gold
df_pessoas.to_parquet(pasta_completa, index=False)
print(f"Arquivo salvo em: {pasta_completa}")
```

```python
### 3.2 - Dimensão Condição Pagamento
###       3.2.1 - Definir caminhos de leitura e gravação de arquivos relacionados as camadas Silver e Gold
###       3.2.2 - Criar Dimensão Condição Pagamento na camada Gold
###       3.2.3 - Criar Surrogate Key para Dimensão Condição Pagamento
###       3.2.4 - Salvar Dimensão Condição Pagamento em formato parquet

# Lê os dados da camada silver 
subpasta = 'condicao_pagamento'
pasta_completa = os.path.join(silver_path, subpasta)
df_condicao_pagamento = pd.read_parquet(os.path.join(pasta_completa)) 

# Cria chaves substitutas (surrogate keys) 
df_condicao_pagamento['cp_sk'] = range(1, len(df_condicao_pagamento) + 1)

# Salva as dimensões na camada gold 
nome_arquivo = 'condicao_pagamento.parquet'
pasta_completa = os.path.join(gold_path, subpasta, nome_arquivo)

# Cria diretórios caso não existam 
os.makedirs(os.path.dirname(pasta_completa), exist_ok=True)

# Salva Arquivo Parquet na Gold
df_condicao_pagamento.to_parquet(pasta_completa, index=False)
print(f"Arquivo salvo em: {pasta_completa}")
```

```python
### 3.3 - Dimensão Tipo de Documento
###       3.3.1 - Definir caminhos de leitura e gravação de arquivos relacionados as camadas Silver e Gold
###       3.3.2 - Criar Dimensão Tipo de Documento na camada Gold
###       3.3.3 - Criar Surrogate Key para Dimensão Tipo de Documento
###       3.3.4 - Salvar Dimensão PesTipo de Documentosoas em formato parquet

# Lê os dados da camada silver
subpasta = 'tipo_doc'
pasta_completa = os.path.join(silver_path, subpasta)
df_tipo_doc = pd.read_parquet(os.path.join(pasta_completa)) 

# Cria chaves substitutas (surrogate keys) 
df_tipo_doc['td_sk'] = range(1, len(df_tipo_doc) + 1)

# Salva as dimensões na camada gold 
nome_arquivo = 'condicao_pagamento.parquet'
pasta_completa = os.path.join(gold_path, subpasta, nome_arquivo)

# Cria diretórios caso não existam 
os.makedirs(os.path.dirname(pasta_completa), exist_ok=True)

# Salva Arquivo Parquet na Gold
df_tipo_doc.to_parquet(pasta_completa, index=False)
print(f"Arquivo salvo em: {pasta_completa}")
```

```python
### 3.4 - Fato: Fluxo de Caixa
###       3.4.1 - Definir caminhos de leitura e gravação de arquivos relacionados as camadas Silver e Gold
###       3.4.2 - Criar Fato Fluxo de Caixa na camada Gold
###       3.4.3 - Agregar as chaves substitutas na tabela Fato Fluxo de Caixa
###       3.4.4 - Salvar Fato Fluxo de Caixa em formato parquet particionado por Ano/Mês
 
# Lê os dados da camada silver
subpasta = 'transacoes'
pasta_completa = os.path.join(silver_path, subpasta)
df_fato_fcx = pd.read_parquet(os.path.join(pasta_completa)) 
df_fato_fcx
```

```python
# Agregar as chaves substitutas na tabela Fato
df_fato_fcx = (
    df_fato_fcx
    .merge(df_pessoas, on='cod_clifor')
    .merge(df_condicao_pagamento, on='condicao_pagamento')
    .merge(df_tipo_doc, on='tipo_doc')
)

# Seleciona as colunas da tabela fato com as chaves substitutas
df_fato_fcx_gold = df_fato_fcx[['documento', 'pessoa_sk', 'td_sk', 'cp_sk', 'valor', 'data_emissão', 'data_vencimento', 'data_pagamento']]

# Exibe o DataFrame final
df_fato_fcx_gold
```

```python
# Define caminhos 
gold_path = r"C:\Portfolio_Projetos\Engenharia_Dados\Pandas-Python\Fonte Dados\Fluxo de Caixa\gold"
subpasta = 'transacoes'
pasta_completa = os.path.join(gold_path, subpasta)

# Certifique-se de que a coluna 'data_emissão' esteja no formato datetime
df_fato_fcx_gold['data_emissão'] = pd.to_datetime(df_fato_fcx_gold['data_emissão'])

# Adiciona colunas de Ano e Mês
df_fato_fcx_gold['Ano'] = df_fato_fcx_gold['data_emissão'].dt.year
df_fato_fcx_gold['Mes'] = df_fato_fcx_gold['data_emissão'].dt.month

# Salva o arquivo parquet particionado automaticamente
df_fato_fcx_gold.to_parquet(
    pasta_completa,
    index=False,
    partition_cols=['Ano', 'Mes'],
    engine='pyarrow'  # Certifique-se de que o PyArrow está instalado
)

print(f"Arquivo salvo particionado em: {pasta_completa}")
```
